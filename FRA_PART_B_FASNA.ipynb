### 2: Stock Returns Calculation and Analysis
- Calculate Returns for all stocks - Calculate the Mean and Standard Deviation for the returns of all stocks - Draw a plot of Mean vs Standard Deviation for all stock returns - Write observations and inferences


###### Steps for calculating returns from prices:

Take logarithms
Take differences

```python
## Calculate Returns for all stocks 

stocks_return=np.log(df.drop(["Date","dates"],axis=1)).diff(axis=0,periods=1)
```

```python
#Checking the rows & columns of dataset
stocks_return.shape
```

```python
#FIRST 5 ROWS
stocks_return.head()
```

###### Calculate the Mean and Standard Deviation for the returns of all stocks

```python
stock_means = stocks_return.mean(axis = 0)
stock_means
```

```python
# Calculating stock standard deviation
stock_sd = stocks_return.std(axis = 0)
stock_sd
```

```python
df1 = pd.DataFrame({'Average':stock_means, 'Volatility': stock_sd})
df1
```

```python
import seaborn as sns
```

```python
#### Let us plot & see what they say about stock prices(ITC_Limited)
plot = sns.scatterplot(data=df1,x='Volatility',y='Average')
plot.axvline(x=0.035904,linestyle='--', color = "red")
plot.axhline(y=0.001634,linestyle='--', color = "red")
plt.title(" Mean vs Standard Deviation plot of ITC_Limited")
plt.show()
```
